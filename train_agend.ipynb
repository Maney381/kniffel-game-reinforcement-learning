```python
import gym
import numpy as np
from stable_baselines3 import DQN
from yahtzee_env import YahtzeeEnv
```

```python
env = YahtzeeEnv()
```

```python
# Create the model
model = DQN("MlpPolicy", env, verbose=1, learning_rate=1e-3, batch_size=32, buffer_size=50000, gamma=0.99)

# Train the model
model.learn(total_timesteps=100000)  # Train for 100,000 steps
```

```python
model.save("yahtzee_dqn_model")
```

```python
model = DQN.load("yahtzee_dqn_model")
obs = env.reset()

for _ in range(20):  # Play 20 rounds
    action, _states = model.predict(obs)  # Get best action
    obs, reward, done, _ = env.step(action)  # Take action
    env.render()  # Show game state
    if done:
        obs = env.reset()
```
